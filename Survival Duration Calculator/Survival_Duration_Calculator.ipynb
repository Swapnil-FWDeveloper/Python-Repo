```python
def calculate_duration(age, unit):

    seconds_in_minute = 60
    seconds_in_hour = 3600
    seconds_in_day = 86400
    seconds_in_week = 604800
    seconds_in_month = 2628000  
    seconds_in_year = 31536000  

    years = age
    months = age * 12
    weeks = age * 52
    days = age * 365 
    hours = days * 24
    minutes = hours * 60
    seconds = minutes * 60

    if unit.lower() == 'years' or unit.lower() == 'y':
        print("You lived for", years, "Years")
    elif unit.lower() == 'months' or unit.lower() == 'm':
        print("You lived for", months, "Months")
    elif unit.lower() == 'weeks' or unit.lower() == 'w':
        print("You lived for", weeks, "Weeks")
    elif unit.lower() == 'days' or unit.lower() == 'd':
        print("You lived for", days, "Days")
    elif unit.lower() == 'hours' or unit.lower() == 'h':
        print("You lived for", hours, "Hours")
    elif unit.lower() == 'minutes' or unit.lower() == 'min':
        print("You lived for", minutes, "Minutes")
    elif unit.lower() == 'seconds' or unit.lower() == 'sec':
        print("You lived for", seconds, "Seconds")
    else:
        print("Invalid unit selected. Please choose from Years, Months, Weeks, Days, Hours, Minutes, or Seconds.")


age = int(input("What's your age? "))
unit = input("Please choose time unit: ").strip()
calculate_duration(age, unit)
```
